Este comando está criando uma colunas onde consta somente as informações que contém dados em números.

```python
colunas= ["age","study_hours_per_day","social_media_hours","netflix_hours","attendance_percentage","sleep_hours","exercise_frequency","mental_health_rating","exam_score"]
```

Este comando criou um gráfico mapa de calor para identificar qual fator mais impactou nas notas do alunos.

```python
sns.heatmap(data=df[colunas].corr(), annot=True, cmap="coolwarm",fmt=".2f");
```

Este comando criou um gráfico de linha para analisar quantas horas seria necessário para os alunos tirar uma boa nota.

```python
sns. lmplot(data=df,x="study_hours_per_day", y="exam_score");
```

Este comando criou um gráfico com intuito de analisar os impactos que a quantidade de horas impacta nas notas dos alunos.

```python
sns. lmplot(data=df,x="netflix_hours", y="exam_score");
```

Este comando criou um gráfico de linha para analisar a quantidade de horas e a saúde mental dos alunos.

```python
sns. lmplot(data=df,x="study_hours_per_day", y="mental_health_rating");
```

Este comando é um filtro para analisarmos a média de horário que horário precisam para estudar para ter uma nota boa e baixa.

```python
filtro_horas_altas = df["study_hours_per_day"] >= 5
filtro_horas_baixas = df["study_hours_per_day"] < 2

media_alta = df[filtro_horas_altas]["exam_score"]. mean()
media_baixa =df[filtro_horas_baixas]["exam_score"]. mean()

print(f"A média das pessoas que estudaram 5 ou mais horas {media_alta:.2f}.Já a média, das pessoas que estudaram menos de 5 horas, foi {media_baixa:.2f}")
```
