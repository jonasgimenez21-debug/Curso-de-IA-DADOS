No entanto, a diferença é relativamente pequena (cerca de 1 ponto), o que sugere que esse fator tem impacto bem menor do que hábitos como horas de estudo e saúde mental, que mostraram associações muito mais fortes com o desempenho acadêmico.
```
A escolaridade dos pais (parental_education_level) tem relação com o desempenho?
```
Sim, é possível observar diferenças nas médias das notas conforme a escolaridade dos pais (parental_education_level), embora elas não sejam muito grandes nesta base.

| Escolaridade dos pais | Média da nota (`exam_score`) |
| --------------------- | ---------------------------: |
| **Bachelor**          |                    **70,27** |
| **Desconhecido**      |                    **70,03** |
| **High School**       |                    **69,55** |
| **Master**            |                    **68,09** |


```
Existe uma quantidade ideal de horas de sono para tirar as melhores notas?
```
Os dados sugerem que sim, mas é importante analisar faixas de horas de sono, e não valores exatos (como 7,2 ou 8,4 horas), pois alguns valores individuais podem ter poucos alunos.

Média da nota por faixa de horas de sono

| Horas de sono        | Média da nota (`exam_score`) |
| -------------------- | ---------------------------: |
| **Menos de 5 horas** |                    **65,01** |
| **5 a 6 horas**      |                    **69,04** |
| **6 a 7 horas**      |                    **69,39** |
| **7 a 8 horas**      |                    **72,44** |
| **8 a 9 horas**      |                    **70,27** |
| **9 horas ou mais**  |                   **75,70*** |

A faixa de 9 horas ou mais pode conter poucos alunos, portanto esse resultado deve ser interpretado com cautela.

Interpretação
Os alunos que dormem menos de 5 horas apresentam o pior desempenho médio.
As notas aumentam à medida que o tempo de sono se aproxima de 7 a 8 horas.
A faixa de 7 a 8 horas concentra um dos melhores desempenhos e é, provavelmente, a recomendação mais confiável por representar um padrão comum.
Embora a faixa de 9 horas ou mais tenha a maior média, esse resultado pode ser influenciado por um número reduzido de estudantes.
Conclusão para a direção da escola

Os resultados indicam que dormir muito pouco está associado a um desempenho acadêmico inferior. Nesta base, os alunos que dormem em torno de 7 a 8 horas por noite apresentam, em média, melhores resultados, reforçando a importância de incentivar hábitos de sono adequados. O aparente desempenho ainda maior entre alunos que dormem 9 horas ou mais deve ser interpretado com cautela, pois pode refletir uma amostra pequena e não necessariamente um efeito real.

No relatório, crie a seção **"2. Novas análises"** e cole os insights que achou mais interessantes.

**Atenção:** alguns desses fatores quase não mudam a nota. Isso também é uma descoberta, pois mostra que nem tudo que imaginamos ser importante realmente é.
