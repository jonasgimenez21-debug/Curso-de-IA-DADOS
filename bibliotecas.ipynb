```python
#Importação de bibliotecas
import numpy as np # np será o apelido desta biblioteca
```

```python
temperaturas = np.array([30,40,50])
print(temperaturas)
print(type(temperaturas))
```

```python
media =np.mean(temperaturas)
print(media)
```

```python
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
menor_valor = np.min(temperaturas)
print(menor_valor)
```

```python
salarios = np.array([2,3,4,5,1,500])
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
mediana = np.median(salarios)
print(mediana)
```

Desafio final:
- Criar um array de habitantes de 5 cidades do estado de SP
- Maior população
- Menor população
- Média de população
- Mediana 

```python
cidades = np.array([11900,1340,1340,840,780])
maior_população = np.max(cidades)
print ( maior_população)
menor_população = np.min(cidades)
print(menor_população)
media_população = np.mean(cidades)
print(media_população)
mediana_população = np.median(cidades)
print(mediana_população)
```
