Programa que avalia as notas que o usuário do nosso aplicativo inseriu.

```python
avaliacao = int(input("Digite a sua avaliação: 0 - 10 estrelas"))

if avaliacao > 8:
    print("Nota excelente")
elif avaliacao > 6:
    print("Avaliação regular")
elif avaliacao > 4:
    print("Avaliação ruim")
else:
    print("Avaliação péssima")
    
```

Sistema que irá verificar se o filme avaliado é bom ou não.
- Criar uma variavel que irá armazenar o nome do filme
- Criar uma variavel que irá armazenar a nota do filme (0 - 5 )
- Condição: Se possuir 5 estrelas, filme Vencedor do Oscar.
- Se possuir de 3 a 4,99, filme bom para assistir.
- Abaixo disso, filme não indicado

```python
Filme = input("Nome do Filme")
avaliacao = float(input("Digite a sua avaliação: 0 - 5"))
if avaliacao ==5:
    print(f"Filme {Filme }Vencedor de Oscar")
elif avaliacao >3:
    print(f"Filme {Filme} Bom para assistir")
else:
    print(f"Filme {Filme} não indicado")
```

```python
Filme = input("Nome do Filme")
avaliacao = float(input("Digite a sua avaliação: 0 - 5"))
if avaliacao <= 0 or avaliacao > 5 :
    print("❌ Nota invalida")
else:
    if avaliacao ==5:
        print(f"Filme {Filme}Vencedor de Oscar")
    elif avaliacao >3:
        print(f"Filme {Filme} Bom para assistir")
    else:
        print(f"Filme {Filme} não indicado")
    
```
